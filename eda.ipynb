```python
import geopandas as gpd

def shapefile_to_csv(shp_path, csv_path):
    # Read the shapefile using geopandasb
    gdf = gpd.read_file(shp_path)
    
    # Save the attribute data and geometry to a CSV file
    gdf.to_csv(csv_path, index=False)
    print(f"Shapefile data saved to {csv_path}")

# Example usage
shapefile_path = 'data\\edge_locale21_nces_AL\\edge_locale21_nces_AL.shp'
output_csv_path = 'data\\edge_locale21_nces_AL\\edge_locale21_nces_AL.csv'

shapefile_to_csv(shapefile_path, output_csv_path)
```

```python
import matplotlib.pyplot as plt

top_10_male = b01_data.nlargest(11, 'Total:Male:15 to 19 years').sort_values(by='Total:Male:15 to 19 years', ascending=False)
top_10_male.drop(index=0, inplace=True)

top_10_female = b01_data.nlargest(11, 'Total:Female:15 to 19 years').sort_values(by='Total:Female:15 to 19 years', ascending=False)
top_10_female.drop(index=0, inplace=True)

# Create the bar plot
fig, axes = plt.subplots(1, 2, figsize=(11, 6))
axes[0].bar(top_10_male['Geographic Area Name'], top_10_male['Total:Male:15 to 19 years'], color='skyblue')
axes[0].set_ylabel('Population', fontsize=12)
axes[0].tick_params(axis='x', rotation=45)

axes[1].bar(top_10_female['Geographic Area Name'], top_10_female['Total:Female:15 to 19 years'], color='lightpink')
#axes[1].set_ylabel('Population', fontsize=12)
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.suptitle('Top 10 Largest Black Populations by County \n 15 to 19 Yr Age Group (Male)', fontsize = 15)
#plt.title('15 to 19 Yr Age Group (Male)')
plt.show()
```

```python
import seaborn as sns
```

```python
top_10_all = b01_data.sort_values(by='Total:15 to 19 years', ascending=False).drop(0).head(10)
top_10_all = top_10_all[['Geographic Area Name', 'Total:Male:15 to 19 years','Total:Female:15 to 19 years']]
top_10_all.rename({'Total:Male:15 to 19 years':'Male Population', 'Total:Female:15 to 19 years':'Female Population'}, axis=1, inplace=True)

top_10_all_melted = top_10_all.melt(id_vars='Geographic Area Name', var_name='Group', value_name='Value')
custom_colors = ['skyblue', 'lightpink'] 

plt.figure(figsize=(8, 6))
sns.barplot(x='Geographic Area Name', y='Value', hue='Group', data=top_10_all_melted, palette=custom_colors)
plt.xlabel('County Name', fontsize=10)
plt.ylabel('Population', fontsize=10)
plt.title('Black Population in 10 Largest Counties \n (15 to 19 Yr Age Group)', fontsize=12)
plt.xticks(fontsize = 9, rotation=45)

plt.tight_layout()
plt.savefig('figures\\black_population_15_19.png')
```
