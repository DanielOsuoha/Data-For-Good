```python
county_map['fips_code'] = county_map['fipscty'].astype(str).str.zfill(5)


nonprofit = nonprofit.merge(county_map[['fips_code', 'county name']], on='fips_code', how='left')

nonprofit = nonprofit.rename(columns={'county name': 'county'})

print(nonprofit[['fipstate', 'fips_code', 'county']].head())
```

```python
county_map['fipscty'] = county_map['fipscty'].astype(str)

county_map['fipscty'] = county_map['fipscty'].apply(lambda x: x.zfill(5))
print(county_map)
```

```python
nonprofit = nonprofit.merge(county_map, on='fipscty', how='left')

print(nonprofit)
```

```python
filtered_df = nonprofit[['fipstate_mapped', 'county', 'naics', 'est']]

print(filtered_df)

```

```python
top_counties = filtered_df.groupby(['fipstate_mapped', 'county', 'naics'])['est'].sum().reset_index()

top_10_edu_counties = top_counties[top_counties['naics'] == "Educational Services"]
top_10_edu_counties = top_10_edu_counties.sort_values(by='est', ascending=False).head(10)

top_10_edu_counties
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12, 8))
sns.barplot(x='est', y='county', hue='naics', data=top_10_edu_counties, palette='viridis')

plt.title('Top 10 Counties by Number of Establishments', fontsize=16)
plt.xlabel('Number of Establishments (est)', fontsize=12)
plt.ylabel('County', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
top_countiescsv = filtered_df.groupby(['fipstate_mapped', 'county', 'naics'])['est'].sum().reset_index()

edu_counties = top_countiescsv[top_countiescsv['naics'] == "Educational Services"]

edu_counties_sorted = edu_counties.sort_values(by='est', ascending=False)

edu_counties_sorted.to_csv('educational_services_counties_sorted.csv', index=False)

```
